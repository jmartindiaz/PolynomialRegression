# Polynomial Regression

### Here we will practice the exercises proposed by W3Schools

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
# Axis x: hours of the day
x = []
for i in range (0,24):
    x.append(i)
print (x)
```

```python
# Axis y: speeds
y = []
for i in range(0,24):
    y.append(rd.randrange(50, 200, 5))
print (y)
```

```python
# Graph
plt.scatter(x, y)
```

```python
# Creating model
mymodel = np.poly1d(np.polyfit(x,y,12))
# np.poly1d creates a Polinomial Function as an array. This variable mymodel, if we aggregate a parenthesis with an another variable
# numpy will interprete this variable as the dependent variable of this polynomial 
# np.polyfit creates a polynomial fit beginning with two arrays.
# The third parameter of this method is the iterations number? Is the number of terms?
myline = np.linspace(0,23,100) 
# The function linspace creates a line beginning with a start value, an stop value
# and a number of points to indicate the line precission. Obviusly, we have more
# than theses parameters but we will just use theses.

# If the third parameter is more small, then the line in the graph will be
# more straight.
print(type(mymodel)) # Here we can see that 'mymodel' is a polynomial
```

```python
# We graph it...
plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

# How can I know if my model is correctly fitting the values?

It is important measure how well is the relationship between the values of the x- and y-axis. The relationship is measured with a value called the **r-squared**.

The **r-squared** value is between 0 and 1. If is equal to 0, it means No Relationship, and if is equal to 1 we have a 100% relationship. This value is very important to make predictions.

Python have a ScikitLearn module that allows calculate this value:

```python
from sklearn.metrics import r2_score

print(r2_score(y, mymodel(x)))
```

Our relationship is not good

# Predict Future Values


Now, we can use the information to predict future values. We will still using the same array 'mymodel' and this time we will try to predict the speed of a car that is sensed at a specific location at 17 hs (or 5 PM).

```python
speed = mymodel(17)
print(f'At 5 PM, based on our model, we could predict that this car moves with an speed of {round(speed,2)} km/h')
```

```python
plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.plot(17 , speed , 'ro')
plt.grid(True)
plt.show()
```

In red, we can see the value predicted previously
